## 1. Pandas를 사용한 데이터 전처리

간단한 데이터 전처리 과정을 따라해 봅시다

```python
df = pd.read_csv("유상증자.csv", index_col=0)
df.head(10)
```

```python
# 우선 NaN값을 드롭한다
df = df.dropna(axis=0)
df.head(10)
```

```python
# 회사명이 없음이라고 되있는 것을 제거해보자
df = df.loc[df['회사명'] != "없음"]
df.head(10)
```

```python
df['보고서명'].unique()
```

```python
# 종속회사란 단어가 들어가는 데이터는 모두 제거한다
df = df.loc[~df['보고서명'].str.contains("종속회사")]
df
```

```python
# 잘 제거 되었는지 확인해보자
df.loc[df.보고서명.str.contains("종속회사")]
```

```python
# 공시일을 datetime으로 만들어 index로 변환한다
df.index = pd.to_datetime(df['공시일'])
```

```python
# 공시일이 여전히 값에 남아있으니 제거해준다
df = df.drop("공시일", axis=1)
df
```

```python
# 날짜가 뒤집혀 있으니 정렬해보자
df = df.sort_index()
df.head(10)
```

```python
# 인덱스를 datetime으로 바꿔줬기 때문에 날짜를 기준으로 값을 가져올 수 있다
df.loc["2022-10"]
```

```python
df.loc["2022-10-15":"2022-10-20"]
```

## 2. Pandas의 Apply 

판다스를 사용할 때, 반복문을 사용하는 것은 좋지 않다 (매우 느리다)

대신, Apply 함수를 사용!

```python
# "보고서명" 컬럼의  '제3자'를 '제4자'로 바꿀 수 있을까?
df
```

```python
# 반복문과 조건문을 사용해서 할 수 있다(나쁜 방법)
for col in df["보고서명"]:
    if "제3자" in col:
        col = col.replace("3","4")
        #print(col)
```

```python
# Pandas에는 이런 경우를 위해 apply 함수가 있다!
df['보고서명'].apply(lambda x: x.replace("3","4"))
```

## 3. 변수설정과 참조

깊은 복사 vs 얕은 복사

```python
# 사용할 데이터셋을 불러오겠습니다
from seaborn import load_dataset

# 설명을 위해 데이터셋의 일부만을 사용합니다
df = load_dataset('penguins').head()
df
```

```python
# df2라는 변수를 생성합니다
df2 = df

df2
```

```python
# df의 0,0번째 값을 A로 변경해보겠습니다
df.iloc[0,0] = "A"
df
```

```python
# 앗! df2의 0,0번째 값도 A로 바뀌었습니다!!
df2

# 왜 그런 걸까요?? 
# -> 우리가 처음에 불러온 데이터셋은 컴퓨터 "메모리"의 특정 부분에 임시적으로 저장됩니다. 
# -> df라고 우리가 지정한 변수 이름은 그 메모리의 특정 부분을 "가리키고" 있는 것 입니다.
# -> df2라는 변수명도 메모리의 동일한 부분을 가리키고 있는 것 입니다. 따라서, 우리가 df.iloc[0,0]="A"로 메모리에 위치한 데이터값을 변화를 주자 df2의 값도 변하게 된 것 입니다
## 이 변수 참조 개념은 파이썬에서 정말 중요한 개념입니다!
```

```python
# 이러한 현상을 막으려면 어떻게 해주면 될까요? 
# -> 판다스에선 copy() 함수를 통해서 이 문제를 해결할 수 있습니다
df3 = df.copy()
df3
```

```python
# df3의 변수명을 변경해 보겠습니다
df3.iloc[0,0] = "B"
df3
```

```python
# 이 경우 df의 값은 바뀌지 않은 것을 확인할 수 있습니다!!! -> copy를 해줬기 때문입니다!!!
df
```

```python
# 흔히들 많이 저지르는 컬럼 drop할 때 하는 실수

# 우선 데이터셋을 다시 한번 불러오겠습니다
df = load_dataset("penguins")
df.head()
```
